```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance
samples = 100
```

```python
z = np.random.multivariate_normal(mean, cov, samples)
x, y = z.T
```

```python
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
plt.plot(x / 10, y / 10, 'x')
plt.axis('equal')
plt.show()
```

```python
plt.plot(x / 10, y / 10, 'x')
plt.axis('equal')
plt.show()
```

```python
# Demonstrates that we can make useful representations from random noise
# orig_noise is fraction of original part we retain

def circle_transform(orig_noise = 0):
    t = []
    for i in range(samples):
        norm = np.linalg.norm(z[i])
        t.append((z[i][0] / norm + z[i][0] * orig_noise,  
                  z[i][1] / norm + z[i][1] * orig_noise))
    return np.array(t)
```

```python
x, y = circle_transform(orig_noise=0).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```

```python
x, y = circle_transform(orig_noise=0.2).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
